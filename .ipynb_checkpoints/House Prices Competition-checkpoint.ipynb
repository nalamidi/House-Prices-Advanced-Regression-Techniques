```python
# To drop missing values (NaN)
data = train.select_dtypes(include=[np.number]).interpolate().dropna() 
```

```python
# Since data now represent all our columns with numerical value, and I would like to use all (Numerical columns) but the SalePrice column, I decided to use the drop() in my X to drop both SalePrice and Id column from my data(numeric features). 
# use to transform the y variable for the model
y = np.log(train.SalePrice)
X = data.drop(['SalePrice', 'Id'], axis=1)
```

```python
X.columns
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = .3, random_state = 7)  # This are our polynomial features
```

```python
lr = LinearRegression()
model = lr.fit(X_train, y_train)
print ("R^2 - Value =", model.score(X_test, y_test))
```

```python
predictions = model.predict(X_test)
```

```python
print ('RMSE = ', mean_squared_error(y_test, predictions)) # mean squared error
```

```python
actual_values = y_test
plt.scatter(predictions, actual_values, alpha=.75, color='r') #alpha helps to show overlapping data
plt.title('Linear Regression Model')
plt.xlabel('Predicted Price')
plt.ylabel('Actual Price')
```

```python
for i in range (-2, 3):
    alpha = 10**i
    rm = Ridge(alpha=alpha)
    ridge_model = rm.fit(X_train, y_train)
    preds_ridge = ridge_model.predict(X_test)
    
    plt.scatter(preds_ridge, actual_values, alpha=.75, color='m')
    plt.xlabel('Predicted Price')
    plt.ylabel('Actual Price')
    plt.title('Ridge Regularization with alpha = {}'.format(alpha))
    overlay = 'R^2 =: {}\nRMSE =: {}'.format(
                    ridge_model.score(X_test, y_test),
                    mean_squared_error(y_test, preds_ridge))
    plt.annotate(s=overlay,xy=(12.1,10.6),size='x-large')
```
